```python
import json

with open("airports.json", "r", encoding="utf-8") as f:
    airports_raw = json.load(f)

with open("airlines.json", "r", encoding="utf-8") as f:
    airlines_raw = json.load(f)

print("Airports:", len(airports_raw))
print("Airlines:", len(airlines_raw))
```

```python
clean_airlines = []

for a in airlines_raw:
    iata = a.get("iata", "").strip()
    
    if a.get("active") != "Y":
        continue
    
    if not iata or iata == "-":
        continue
    
    clean_airlines.append({
        "type": "airline",
        "iata": iata.upper(),
        "icao": a.get("icao", "").strip(),
        "name": a.get("name", "").strip(),
        "country": a.get("country", "").strip()
    })

print("Clean airlines:", len(clean_airlines))
```

```python
clean_airports = []

for a in airports_raw:
    iata = a.get("code", "").strip()
    
    if not iata or len(iata) != 3:
        continue
    
    clean_airports.append({
        "type": "airport",
        "iata": iata.upper(),
        "icao": a.get("icao", "").strip(),
        "name": a.get("name", "").strip(),
        "city": a.get("city", "").strip(),
        "country": a.get("country", "").strip()
    })

print("Clean airports:", len(clean_airports))
```

```python
def deduplicate_by_iata(data):
    seen = set()
    unique = []
    for item in data:
        if item["iata"] not in seen:
            unique.append(item)
            seen.add(item["iata"])
    return unique

clean_airports = deduplicate_by_iata(clean_airports)
clean_airlines = deduplicate_by_iata(clean_airlines)

print("Deduplicated airports:", len(clean_airports))
print("Deduplicated airlines:", len(clean_airlines))
```

```python
def deduplicate_by_iata(data):
    seen = set()
    unique = []
    for item in data:
        if item["iata"] not in seen:
            unique.append(item)
            seen.add(item["iata"])
    return unique

clean_airports = deduplicate_by_iata(clean_airports)
clean_airlines = deduplicate_by_iata(clean_airlines)

print("Deduplicated airports:", len(clean_airports))
print("Deduplicated airlines:", len(clean_airlines))
```

```python
with open("clean_airports.json", "w", encoding="utf-8") as f:
    json.dump(clean_airports, f, indent=2, ensure_ascii=False)

with open("clean_airlines.json", "w", encoding="utf-8") as f:
    json.dump(clean_airlines, f, indent=2, ensure_ascii=False)

print("Clean files saved.")
```

```python
import json

with open("airports.json", "r", encoding="utf-8") as f:
    airports = json.load(f)

clean_airports = []

for a in airports:
    if not a.get("code"):
        continue

    try:
        lat = float(a["lat"])
        lon = float(a["lon"])
    except:
        continue

    clean_airports.append({
        "type": "airport",
        "iata": a["code"],
        "icao": a.get("icao", ""),
        "name": a.get("name", ""),
        "city": a.get("city", ""),
        "country": a.get("country", ""),
        "lat": lat,
        "lon": lon,
        "tz": a.get("tz", "").replace("\\/", "/")
    })

with open("clean_airports.json", "w", encoding="utf-8") as f:
    json.dump(clean_airports, f, indent=2)

print("Done. Cleaned:", len(clean_airports))
```
